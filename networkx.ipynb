```python
import networkx as nx
```

```python
g=nx.Graph()
```

```python
g.add_nodes_from(["a","b","c","d"])     
```

```python
g.add_edges_from([("a","b"),("b","c"),("a","c"),("c","d")])
```

```python
g.nodes()
```

```python
g.edges()
```

```python
 g.has_node('a')
```

```python
g.has_edge('a','b')
```

```python
g.number_of_nodes()
```

```python
g.number_of_edges()
```
