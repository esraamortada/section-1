```python
import networkx as nx
```

```python
g=nx.Graph()
```

```python
g.add_edges_from([(1,2),(1,3),(2,3),(3,4)])
```

```python
dict(g.degree())
```

```python
nx.to_numpy_matrix(g)
```

```python
f=nx.read_edgelist('graph.txt',nodetype=str,create_using=nx.DiGraph())
nx.draw(f,with_labels=True,node_size=1000)
```

```python
import pandas as pd
```

```python
df=pd.read_csv('HomoSapiens_binary_hq.txt',nrows=100,sep='\t',usecols=['Gene_A','Gene_B','pmid:method:quality'])
```

```python
df
```

```python
bio_graph=nx.from_pandas_edgelist(df,'Gene_A','Gene_B',create_using=nx.DiGraph())
```

```python
nx.draw(bio_graph,with_labels=True)
```
